```
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("/content/mobile.csv")

plt.figure()
data.groupby('price_range')['ram'].mean().plot(kind='bar')
plt.xlabel("Price Range")
plt.ylabel("Average RAM")
plt.title("Average RAM vs Mobile Price Range")
plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("/content/mobile.csv")

for col in data.columns:
    plt.figure()
    if data[col].nunique() > 10:
        binned = pd.cut(data[col], bins=5)
        binned.value_counts().plot(kind='pie', autopct='%1.1f%%')
    else:
        data[col].value_counts().plot(kind='pie', autopct='%1.1f%%')
    plt.title(f"{col} Distribution")
    plt.ylabel("")
    plt.show()
```

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

data = pd.read_csv("/content/mobile.csv")

values = data['ram']

density = gaussian_kde(values)
xs = np.linspace(values.min(), values.max(), 200)

plt.figure()
plt.hist(values, bins=15, density=True, edgecolor='black')
plt.plot(xs, density(xs))
plt.xlabel("RAM")
plt.ylabel("Density")
plt.title("Histogram of RAM with Density Curve")
plt.show()
```

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

data = pd.read_csv("/content/mobile.csv")

values = data['ram']

density = gaussian_kde(values)
xs = np.linspace(values.min(), values.max(), 200)

bins = pd.cut(values, bins=15)
bar_values = bins.value_counts().sort_index()

plt.figure()
plt.bar(bar_values.index.astype(str), bar_values.values, edgecolor='black')
plt.plot(np.arange(len(bar_values)), density(xs[:len(bar_values)]) * max(bar_values.values))
plt.xticks(rotation=45)
plt.xlabel("RAM Range")
plt.ylabel("Frequency")
plt.title("Bar Graph of RAM with Smooth Curve")
plt.show()
```
